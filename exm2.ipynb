```python
# Cevap1
def kare(a):
    return  a**2

kare(5)
```

```python
def stringlen(a):
    sonuc = len(a)
    return  sonuc

stringlen("Muhammed")
```

```python
def ilklist(a):
    sonuc = a[0]
    return sonuc

ilklist([0,1,2,3])
```

```python
def tekcift(a):
    if a % 2 == 0:
        print("Bu sayı çift")
    elif a % 2 == 1:
        print("Bu sayı Tek")
    else:
        print("Bu sayı geçersizdir.")

tekcift(3)
```

```python
def pozneg(a):
    if a > 0:
        print("Bu sayı pozitifitir")
    elif a == 0:
        print("Bu sayı 0dır.")
    else:
        print("Bu sayı negatiftir.")

pozneg(-1)
```

```python
def  listtop(a):
    if type(a) == list:
        return sum(a)
    else:
        print("Lütfen liste giriniz")

listtop([1,2,3,4,5])
```

```python
def palindrom(a):
    if a == "".join(reversed(a)):
       print("Palindromdur")
    else:
        print("Palindrom değildir.")

palindrom("Kek")
```
