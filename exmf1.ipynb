```python
def tur():# tek çift kontrolü
    a = int(input("Bir sayı giriniz:"))
    if a % 2 == 0:
        return print(f"{a} sayısı çifttir.")
    elif a % 2 == 1:
        return print(f"{a} sayısı tektir.")
    else :
        return print("Lütfen negatif sayı ve karakter girmeyiniz.")

tur()
```

```python
def list_topla():# kullanıcıdan aldığı sayıları toplar
    liste = list(map(int,input("Sayıları aralarında boşluk bırakarak giriniz:").split()))
    return print(f"Sayıların toplamı: {sum(liste)}")

list_topla()
```

```python
# bir sayı listesindeki en büyük ve en küçük sayıyı bulan fonksiyon

def en():
    liste = list(map(int,input("Sayıları aralarında boşluk bırakarak giriniz:").split()))
    enb = liste[0]
    enk = liste[0]
    for eleman in liste:
        if eleman > enb: 
            enb = eleman
        elif eleman < enk:
            enk = eleman
    print(f"En küçük sayı: {enk}, En büyük sayı: {enb}")

en()
```

```python
# Palindrom string kontrolü yapan fonksiyon

def palindrom():
    string = list(map(str,input("Bir kelime giriniz(Aralarına virgül koyunuz):").split(",")))
    palindromlar = []
    not_palindrom = []
    for kelime in string:
        if kelime == kelime[::-1]:
            palindromlar.append(kelime)
        else:
            not_palindrom.append(kelime)
    return print(f"Palindrom kelimeler: {palindromlar}\nPalindrom olmayan kelimeler:{not_palindrom}")

palindrom()
```

```python
def asal():# asal sayı tespiti
    liste = range(int(input("Bir sayı giriniz:")),int(input("Bir sayı giriniz:")))
    asall = []
    not_asal = []
    for i in liste:
        if i > 1:# 1 ve altı asal sayı değildir.
            for x in range(2,i):
                if i % x == 0:
                    not_asal.append(i)
                    break
            else:
                asall.append(i)
    print(f"Asal sayılar:{asall},\nAsal olmayan sayılar:{not_asal}")

asal()
        
```

```python
# Bir cümlede bir kelime kaç defa geçiyor

def kelime_frekans():
    liste = map(str,input("Bir cümle giriniz").split())
    frekans = dict()
    for kelime in liste:
        k = kelime.lower()
        if k in frekans:
            frekans[k] += 1
        else:
            frekans[k] = 1
    return frekans

kelime_frekans()

```

```python
def tersiz():# tekrarsız keliemeleri tespit etme 
    liste = list(map(str,input("Cümleyi giriniz.").split()))
    harfler = []
    tekrarsiz = set()
    for kelime in liste:
        for harf in kelime:
            harfler.append(harf)
            tekrarsiz.add(harf.lower())
    print(f"toplamda cümlede tekrarsız {len(tekrarsiz)} harf vardır.")

tersiz()
```
