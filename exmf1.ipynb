```python
def tur():
    a = int(input("Bir sayı giriniz:"))
    if a % 2 == 0:
        return print(f"{a} sayısı çifttir.")
    elif a % 2 == 1:
        return print(f"{a} sayısı tektir.")
    else :
        return print("Lütfen negatif sayı ve karakter girmeyiniz.")

tur()
```

```python
def list_topla():# kullanıcıdan aldığı sayıları toplar
    liste = list(map(int,input("Sayıları aralarında boşluk bırakarak giriniz:").split()))
    return print(f"Sayıların toplamı: {sum(liste)}")

list_topla()
```

```python
# bir sayı listesindeki en büyük ve en küçük sayıyı bulan fonksiyon

def en():
    liste = list(map(int,input("Sayıları aralarında boşluk bırakarak giriniz:").split()))
    enb = liste[0]
    enk = liste[0]
    for eleman in liste:
        if eleman > enb: 
            enb = eleman
        elif eleman < enk:
            enk = eleman
    print(f"En küçük sayı: {enk}, En büyük sayı: {enb}")

en()
```

```python
# Palindrom string kontrolü yapan fonksiyon

def palindrom():
    string = list(map(str,input("Bir kelime giriniz(Aralarına virgül koyunuz):").split(",")))
    palindromlar = []
    not_palindrom = []
    for kelime in string:
        if kelime == kelime[::-1]:
            palindromlar.append(kelime)
        else:
            not_palindrom.append(kelime)
    return print(f"Palindrom kelimeler: {palindromlar}\nPalindrom olmayan kelimeler:")
```
